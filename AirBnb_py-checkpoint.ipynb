# Mechine Learning


# Start with Linear Regression

```python
dataset_onehot1 = pd.get_dummies(df, columns=['neighbourhood_group',"room_type"], prefix = ['ng',"rt"],drop_first=True)
dataset_onehot1.drop(["neighbourhood"], axis=1, inplace=True)
X1= dataset_onehot1.loc[:, dataset_onehot1.columns != 'price']
Y1 = dataset_onehot1["price"]
x_train1, x_test1, y_train1, y_test1 = train_test_split(X1, Y1, test_size=0.20, random_state=42)
```

```python
dataset_onehot1.shape
```

```python
### Fitting Linear regression
reg1 = LinearRegression().fit(x_train1, y_train1)
```

```python
### R squared value
reg1.score(x_train1, y_train1)
```

```python
# let's look inside the model
print("b1:",reg1.coef_)
print("b0:",reg1.intercept_)
```

```python
### Predicting 
y_pred1 = reg1.predict(x_test1)
```

```python
Coeff1 = pd.DataFrame(columns=["Variable","Coefficient"])
Coeff1["Variable"]=x_train1.columns
Coeff1["Coefficient"]=reg1.coef_
Coeff1.sort_values("Coefficient")
```

As can be seen, there are alot outlires existing so we are cleaning the data for each room type separately

```python
df2=df[df['room_type'] =='Private room'][df["price"]<180][df["price"]>20]
df3=df[df['room_type'] =='Shared room'][df["price"]<120][df["price"]>10]
df4=df[df['room_type'] =='Entire home/apt'][df["price"]>40][df["price"]<230]
df5=df[df['room_type'] =='Hotel room'][df["price"]<200]
df2=df2.append(df3, ignore_index = True)
df2=df2.append(df4, ignore_index = True)
df2=df2.append(df5, ignore_index = True)
df=df2
sns.violinplot(df["room_type"],df["price"])
```

looking better, there are less oulier.

```python
dataset_onehot2 = pd.get_dummies(df, columns=['neighbourhood_group',"room_type"], prefix = ['ng',"rt"],drop_first=True)
dataset_onehot2.drop(["neighbourhood"], axis=1, inplace=True)
X2= dataset_onehot2.loc[:, dataset_onehot1.columns != 'price']
Y2 = dataset_onehot2["price"]
x_train2, x_test2, y_train2, y_test2 = train_test_split(X2, Y2, test_size=0.20, random_state=42)
### Fitting Linear regression
reg2 = LinearRegression().fit(x_train2, y_train2)
```

```python
### R squared value
reg2.score(x_train2, y_train2)
```

```python
# let's look inside the model
print("b1:",reg2.coef_)
print("b0:",reg2.intercept_)
```

```python
### Predicting 
y_pred2 = reg2.predict(x_test2)

Coeff2 = pd.DataFrame(columns=["Variable","Coefficient"])
Coeff2["Variable"]=x_train2.columns
Coeff2["Coefficient"]=reg2.coef_
Coeff2.sort_values("Coefficient")
```

```python
df_pred2 = pd.DataFrame({'Actual': y_test2, 'Predicted': y_pred2})
df_pred2
```
